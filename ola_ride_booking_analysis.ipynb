```python
# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
from sqlalchemy import create_engine
from urllib.parse import quote_plus


username = "postgres" # default username for PostgreSQL
# password = "Sumeet@123" # the password you set during installation
password = quote_plus("Sumeet@123") 
host = "localhost" # if running local
port = 5432 # default PostgreSQL port 
database = "ola" # the database you created in pgAdmin4

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

## Step 2: Load DataFrame into PostgreSQL
table_name = "bookings" # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully Loaded into table '{table_name}' in database '{database}'.")
```

```python
from sqlalchemy import create_engine

# MySQL connection
username = "root"
password = "root"
host = "localhost"
port = "3306"
database = "ola"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Write DataFrame to MySQL
table_name = "bookings"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

# Read back sample
pd.read_sql("SELECT * FROM bookings LIMIT 5;", engine)
```
